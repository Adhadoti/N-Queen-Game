```python
def extract_and_print(read):
    
    with read as in_file:
        content=read.read()
        sent=re.findall("From: (\S+@\S+)",content)
        to=re.findall("To: (\S+@\S+)",content)
        num1=re.findall("X-DSPAM-Confidence: (0.+[0-9])",content)
        num2=re.findall("X-DSPAM-Processed: (.+)",content)
        
        for i in range(0,len(to)):
            if float(num1[i])<=0.80:
                time=dateparser.parse(num2[i])
                num2_final=time.strftime('%m%d%y %H:%M:%S')
                
                print('{:35s}{:38s}{:.2f}     {:40s}'.format(sent[i],to[i],float(num1[i]),num2_final))
                
        
                
    #print('terminate')
        
        
def main():
    
    in_file=open("mbox.txt","r")
    print("-----------------------------------------------------------------------------------------------------------")
    print('{:35s}{:35s}{:20s}{:30s}'.format('To', 'From','X-DSPAM-Confidence','X-DSPAM-Processed'))
    print("-----------------------------------------------------------------------------------------------------------")
    extract_and_print(in_file)
    
main()
```
